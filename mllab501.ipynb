```python
#1 Question
import numpy as np
import matplotlib.pyplot as plt
#  the step activation function
def step(x):
    return np.where(x >= 0, 1, 0)
#the AND gate training data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 1, 1])
# the initial weights and learning rate are
w = np.array([8, 0.8, -0.25])
alpha = 0.05
error_threshold = 0.002
max_epochs = 100
# Initializing the error and epoch counters
errors = []
epochs = 0
# Train the perceptron
while True:
# Calculating the net input
    net_input = np.dot(X, w[1:]) + w[0]
#Apply the activation function
    output = step(net_input)
    error = y - output
# Calculating the sum-squared error
    sse = np.sum(error ** 2)
    errors.append(sse)
    epochs += 1
    w[1:] += alpha * np.dot(X.T, error)
    w[0] += alpha * error.sum()
    if sse <= error_threshold or epochs >= max_epochs:
        break
# Plot the errors
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Sum-squared error')
plt.show()
# Print the number of epochs needed for convergence
print('Number of epochs:', epochs)
```
